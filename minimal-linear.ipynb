```python
import cPickle as pickle
ft_table = []
bt_table = []
for n_layers in range(100, 1100, 100):
    path = 'info/batch_size-1-in_features-1-n_iterations-100-n_layers-%d-out_features-1' % n_layers
    ft, bt = pickle.load(open(path))
    ft_table.append(ft)
    bt_table.append(bt)
```

```python
import matplotlib.pylab as pl
%matplotlib inline
```

```python
pl.plot(range(100, 1100, 100), ft_table, label='forward')
pl.plot(range(100, 1100, 100), bt_table, label='backward')
pl.legend(framealpha=0, loc=0)
pl.xlabel('number of layers')
pl.ylabel('time (s)')
```
