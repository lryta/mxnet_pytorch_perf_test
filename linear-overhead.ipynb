```python
import cPickle as pickle

ft_table = []
bt_table = []
f_range = tuple(2 ** i for i in range(12))
for f in f_range:
    path = 'info/batch_size-256-in_features-%d-n_iterations-100-n_layers-100-out_features-%d' % (f, f)
    ft, bt = pickle.load(open(path))
    ft_table.append(ft)
    bt_table.append(bt)
```

```python
import matplotlib.pylab as pl
%matplotlib inline
```

```python
pl.plot(f_range, ft_table, label='forward')
pl.plot(f_range, bt_table, label='backward')
pl.plot(f_range, (0.01,) * len(f_range))
pl.legend(framealpha=0, loc=0)
pl.xlabel('number of hidden units')
pl.ylabel('time (s)')
```
